```python
import os
import pandas as pd
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
```

```python
#if has HTTP error, update the address for the cvs file in github
fp_job_2019 = 'https://raw.githubusercontent.com/COGS108/Group065Sp22/master/data/occupation_2019.csv?token=GHSAT0AAAAAABTENZP5FJPUPLTBQIHFMV32YTTMJAA'
job_2019 = pd.read_csv(fp_job_2019)
```

```python
#quick glance of the data set
job_2019
```

```python
#Tranform blank data to NaN
job_2019 = job_2019.replace('–', np.NaN)

#Reset occupation as index
job_2019.index = job_2019['occupation']

#Create DF for women
women_job_2019 = job_2019[job_2019.columns[job_2019.columns.str.contains('_w')]]
women = pd.Series(['women'] * women_job_2019.shape[0])
women.name = 'gender'
women.index = women_job_2019.index
women_job_2019 = women_job_2019.merge(women, how = 'outer', left_index = True, right_index = True, copy = False)

#Create DF for men
men_job_2019 = job_2019[job_2019.columns[job_2019.columns.str.contains('_m')]]
men = pd.Series(['men'] * men_job_2019.shape[0])
men.name = 'gender'
men.index = men_job_2019.index
men_job_2019 = men_job_2019.merge(men, how = 'outer', left_index = True, right_index = True, copy = False)

#Re-join DF for women and men to new, cleaned data
job_2019 = pd.concat([women_job_2019, men_job_2019], join = 'outer')

#quick glance of the cleaned DF
job_2019
```

```python
#reorder and rename the columns
new_idx = ['num_workers_w','median_week_earnings_w','se_of_median_w','gender',
           'num_workers_m','median_week_earnings_m','se_of_median_m',]
job_2019.columns = new_idx
job_2019 = job_2019.reindex(columns=[new_idx[3]]+ new_idx[0:3]+new_idx[4:])
```
